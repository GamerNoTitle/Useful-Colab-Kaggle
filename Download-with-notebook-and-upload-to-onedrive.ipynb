```python
#@title Setting the config of rclone
import os
import wget
from urllib.parse import unquote
from rclone.rclone import Rclone
rc = Rclone()
config = """
"""
with open('/root/.config/rclone/rclone.conf', 'wt') as f:
  f.write(config)
  f.close()
!cat /root/.config/rclone/rclone.conf # Check whether the config has been added successfully.

# If you can see the config below, then you have successfully added the config to rclone.
```

```python
#@title Set the target cloud
rc.unit = 'MB'
# drive = "Please input your drive name here" #@param {type:"string"}
# path = "Please input the path your your target drive here (e.g. /download)" #@param {type: "string"}
drive = 'default'
path = '/download'
```

```python
#@title (Optional) Clear target folder [Currently useless]
rc.execute(f'rmdir -f {drive}:{path}')
rc.execute(f'mkdir {drive}:{path}')
```

```python
#@title Add your links here, download & Upload
# e.g.
# [
#     ('https://bili33.top/', "file"),
#     ('https://example.com/file', "filename")
# ]
# If you left the filename empty, it will be generated automatically
links = [

] 
%cd download
!rm -rf *
for link, file in links:
  if file == '':
    file = unquote(link.split('/')[-1])
    link = unquote(link)
    print(f'Downloading {file} from {link}')
    file = wget.download(link)
    print(f'Moving {file} to {drive}:{path}')
    rc.copy(file, f'{drive}:{path}')
    print(f'\nDeleting {file} from disk...')
    os.remove(file)
    print('Done')
  else:
    file = unquote(file)
    print(f'Downloading {file} from {link}')
    wget.download(link, file)
    print(f'Moving {file} to {drive}:{path}')
    rc.copy(file, f'{drive}:{path}')
    print(f'\nDeleting {file} from disk...')
    os.remove(file)
    print('Done')
```
